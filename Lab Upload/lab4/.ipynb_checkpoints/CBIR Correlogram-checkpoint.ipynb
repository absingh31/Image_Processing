## Instantiating CBIR Engine

```python
from CBIR_Correlogram import CBIR_Correlogram
```

**Indexing the Database and saving it to disk**


```python
engine = CBIR_Correlogram(database='imageDB/', save_index=True) 
```

**Loading indexed database from disk**

```python
engine = CBIR_Correlogram(database='imageDB/', use_index='index')
```

## Querying Images

```python
engine.query(query='imageDB/1.jpg', rank=10)
```

## Evaluating Performace with Precision & Recall

```python
engine.query(query='imageDB/1.jpg', rank=3, precision_recall=True,
             relevant=['1.jpg', '2.jpg', '3.jpg', '4.jpg', '5.jpg', '6.jpg', '7.jpg', '8.jpg', '9.jpg', '10.jpg',])
```
