```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Image loaded in grayscale
img = cv2.imread('back_pose.jpg',0) 
'''   0 for grayscale
     1 for color with alpha
    -1 for %color      '''
```

```python
cv2.imshow('dude',img)   # replace anything with title
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# showing using matplotlib

plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.plot([50,100],[80,100],'c', linewidth=5)
plt.show()
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    cv2.imshow('frame', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
